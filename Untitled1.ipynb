```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('vehicles_dataset_upd.csv')
df.head()
```

```python
df_with_out = pd.read_csv('df_types_upd.csv')
df_with_out.head()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df.odometer,bins=20,color='purple')
plt.xlabel('Кол-во пройденных миль')
plt.ylabel('Частота значения')
plt.title('Распределение количества пройденных миль')
```

```python
plt.figure(figsize=(8,5))
plt.hist(df_with_out.odometer,bins=100,color='purple')
plt.xlabel('Кол-во пройденных миль')
plt.ylabel('Частота значения')
plt.title('Распределение количества пройденных миль')
```
